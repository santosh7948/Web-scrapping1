```python
import requests 
from bs4 import BeautifulSoup
import  pandas as pd
quotes=[]
for pageNumber in range(1,5):
    url=f"http://quotes.toscrape.com/page/{pageNumber}/"
    response=requests.get(url)
    response = response.content
    soup = BeautifulSoup(response,'html.parser')
    div=soup.find('div')
    contents = soup.find_all('div', class_='quote')
    for q in contents:
        quote={}
        quote["quate"]=q.find('span', class_='text').text
        quote["author"]=q.find('small', class_='author').text   #original tags
        tags=q.find('div', class_='tags')
        quote["orgtag"]=tags.find('meta', class_='keywords')['content']
        quotes.append(quote)
quotes
    
```

```python
df=pd.DataFrame(quotes)
```
